```python
import os
import pandas as pd
import plotly.express as px
import hvplot.pandas
from pathlib import Path
from dotenv import load_dotenv
%matplotlib inline
import numpy as np
```

```python
# Timeperiod: 2017-2021 S&P500  DIGITAL ASSETS: BTC, ETH, LTC, XLM, XRP, ZEC
crypto_2021 = pd.read_csv(
    Path('./2021_crypto_data.csv'),
    parse_dates=True,
    infer_datetime_format=True,
    index_col='time')
    
```

```python
crypto_2021.columns
```

```python
crypto_2021
```

```python
round(crypto_2021.describe(),2)
```

```python
# Daily Returns
btc_daily_returns = crypto_2021['btc_price_close'].pct_change().dropna()
eth_daily_returns = crypto_2021['eth_price_close'].pct_change().dropna()
ltc_daily_returns = crypto_2021['ltc_price_close'].pct_change().dropna()
zec_daily_returns = crypto_2021['zec_price_close'].pct_change().dropna()
usdt_daily_returns = crypto_2021['usdt_price_close'].pct_change().dropna()
xlm_daily_returns = crypto_2021['xlm_price_close'].pct_change().dropna()
xrp_daily_returns = crypto_2021['xrp_price_close'].pct_change().dropna()
dash_daily_returns = crypto_2021['dash_price_close'].pct_change().dropna()
ada_daily_returns = crypto_2021['ada_price_close'].pct_change().dropna()
sol_daily_returns = crypto_2021['sol_price_close'].pct_change().dropna()
doge_daily_returns = crypto_2021['doge_price_close'].pct_change().dropna()
lisk_daily_returns = crypto_2021['lisk_price_close'].pct_change().dropna()
waves_daily_returns = crypto_2021['waves_price_close'].pct_change().dropna()
link_daily_returns = crypto_2021['link_price_close'].pct_change().dropna()
```

```python
daily_returns_df = pd.concat(
    [btc_daily_returns,eth_daily_returns,ltc_daily_returns,zec_daily_returns,usdt_daily_returns,xlm_daily_returns,xrp_daily_returns,dash_daily_returns,ada_daily_returns,sol_daily_returns,doge_daily_returns,lisk_daily_returns,waves_daily_returns,link_daily_returns],axis=1, join='inner')
```

```python
daily_returns_df.index = pd.to_datetime(daily_returns_df.index)
```

```python
daily_returns_df.index = daily_returns_df.index.date
```

```python
daily_returns_df
```

```python
daily_returns = daily_returns_df.rename(columns = {'btc_daily_returns':'btc_daily_returns','eth_price_close':'eth_daily_returns','ltc_price_close':'ltc_daily_returns','zec_price_close':'zec_daily_returns','usdt_price_close':'usdt_daily_returns','xlm_price_close':'xlm_daily_returns','xrp_price_close':'xrp_daily_returns','dash_price_close':'dash_daily_returns','ada_price_close':'ada_daily_returns','sol_price_close':'sol_daily_returns','doge_price_close':'doge_daily_returns','lisk_price_close':'lisk_daily_returns','waves_price_close':'waves_daily_returns','link_price_close':'link_daily_returns'})
```
