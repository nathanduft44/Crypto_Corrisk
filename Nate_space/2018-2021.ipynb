```python
#Importing Libraries
import pandas as pd 
import datetime 
import numpy as np
import scipy.stats
# Download pandas data reader with conda install pandas-datareader
import pandas_datareader.data as web
import matplotlib.pyplot as plt
from pandas import DataFrame
import itertools
from math import sqrt
from scipy.stats import norm
#For optimization
import timeit
%matplotlib inline
import hvplot.pandas
```

```python
#Start date and End date is refers to time period for fetching the data using the Pandas dataframe API
#We have selected data for 2020 up to current time 
start_date = datetime.datetime(2018, 8, 18)
end_date = datetime.datetime(2021, 8, 18)

#Source is yahoo finance
source = 'yahoo'
```

```python
#We read each ticker and download information from start to end date
all_data = {ticker:web.DataReader(ticker,source,start_date,end_date)
            for ticker in ["BTC-USD", "ETH-USD", "LTC-USD","XLM-USD", "DOGE-USD","XRP-USD","ADA-USD","LINK-USD"]}
```

```python
#We store only Adjusted Close price in a seperate dataset for analysis
crypto_price = pd.DataFrame({ticker: data['Adj Close']
                                  for ticker, data in all_data.items()})
```

```python
crypto_price.describe()
```

```python
# Daily Returns
daily_returns = crypto_price.pct_change().dropna() 
daily_returns.rename(columns = {"BTC-USD":"BTC Daily Returns","ETH-USD":"ETH Daily Returns","LTC-USD":"LTC Daily Returns","XLM-USD":"XLM Daily Returns","DOGE-USD":"DOGE Returns","XRP-USD":"XRP Daily Returns","ADA-USD":"ADA Daily Returns","LINK-USD":"LINK Daily Returns"})
```

```python
"""
SHARPE RATIO ANALYSIS
Sharpe ratio = Annualized average return รท Annualized standard deviation
"""
```

```python
#Annualized Average Returns
annualized_avg_annual_returns = daily_returns.mean() * 365
# Annualized Standard Deviation
average_std_dev_cryptos = daily_returns.std() * np.sqrt(365)
annualized_avg_annual_returns
```

```python
average_std_dev_cryptos
```

```python
## Calulations for Sharpe Ratio
```

```python
btc_sharpe = annualized_avg_annual_returns['BTC-USD'] / average_std_dev_cryptos['BTC-USD']
eth_sharpe = annualized_avg_annual_returns['ETH-USD'] / average_std_dev_cryptos['ETH-USD']
ltc_sharpe = annualized_avg_annual_returns['LTC-USD'] / average_std_dev_cryptos['LTC-USD']
xlm_sharpe = annualized_avg_annual_returns['XLM-USD'] / average_std_dev_cryptos['XLM-USD']
doge_sharpe = annualized_avg_annual_returns['DOGE-USD'] / average_std_dev_cryptos['DOGE-USD']
xrp_sharpe = annualized_avg_annual_returns['XRP-USD'] / average_std_dev_cryptos['XRP-USD']
ada_sharpe = annualized_avg_annual_returns['ADA-USD'] / average_std_dev_cryptos['ADA-USD']
link_sharpe = annualized_avg_annual_returns['LINK-USD'] / average_std_dev_cryptos['LINK-USD']
```
