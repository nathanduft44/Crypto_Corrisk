```python
# Import the required libraries and dependencies
import numpy as np
from pathlib import Path
import pandas as pd
%matplotlib inline 
import seaborn as sns
```

```python
# Import the data by reading in the CSV file and setting the DatetimeIndex
# Review the  rows of the DataFrame
df_2019 = pd.read_csv(
    Path("2019_data.csv"),
)

```

```python
df_2019.describe()
```

```python
df_2019.columns
```

```python
final_2019_df = pd.DataFrame()
final_2019_df['BTC_Monthly_Close'] = df_2019['price_close']
final_2019_df['ETH_Monthly_Close'] = df_2019['price_close.1']
final_2019_df['LTC_Monthly_Close'] = df_2019['price_close.2']
final_2019_df['ADA_Monthly_Close'] = df_2019['price_close.3']
final_2019_df['XLM_Monthly_Close'] = df_2019['price_close.4']
final_2019_df['XRP_Monthly_Close'] = df_2019['price_close.5']
final_2019_df['ZEC_Monthly_Close'] = df_2019['price_close.6']
final_2019_df['USDT_Monthly_Close'] = df_2019['price_close.7']
final_2019_df['DASH_Monthly_Close'] = df_2019['price_close.8']

final_2019_df

```

```python
# calculating onthly returns in dataframe
monthly_returns = final_2019_df.pct_change().dropna()
monthly_returns
```

```python
# calculate bitcoin variance
# assign to market variance value
market_var = monthly_returns['BTC_Monthly_Close'].var()
market_var
```

```python
# Calculate the covariance
market_cov = monthly_returns['BTC_Monthly_Close'].cov(monthly_returns['BTC_Monthly_Close'])
market_cov
```

```python
eth_cov = monthly_returns['ETH_Monthly_Close'].cov(monthly_returns['BTC_Monthly_Close'])
ltc_cov = monthly_returns['LTC_Monthly_Close'].cov(monthly_returns['BTC_Monthly_Close'])
ada_cov = monthly_returns['ADA_Monthly_Close'].cov(monthly_returns['BTC_Monthly_Close'])
xlm_cov = monthly_returns['XLM_Monthly_Close'].cov(monthly_returns['BTC_Monthly_Close'])
xrp_cov = monthly_returns['XRP_Monthly_Close'].cov(monthly_returns['BTC_Monthly_Close'])
zec_cov = monthly_returns['ZEC_Monthly_Close'].cov(monthly_returns['BTC_Monthly_Close'])
usdt_cov = monthly_returns['USDT_Monthly_Close'].cov(monthly_returns['BTC_Monthly_Close'])
dash_cov = monthly_returns['DASH_Monthly_Close'].cov(monthly_returns['BTC_Monthly_Close'])

```

```python
covarience_list = [market_cov, eth_cov, ltc_cov, ada_cov,xlm_cov,xrp_cov,zec_cov,usdt_cov,dash_cov]
covarience_list
```

```python
# calculating BETA

btc_beta = market_cov / market_var
eth_beta = eth_cov / market_var
ltc_beta = ltc_cov / market_var
ada_beta = ada_cov / market_var
xlm_beta = xlm_cov / market_var
xrp_beta = xrp_cov / market_var
zec_beta = zec_cov / market_var
usdt_beta = usdt_cov / market_var
dash_beta = dash_cov / market_var
```

```python
beta_list = [btc_beta, eth_beta,ltc_beta,ada_beta,xlm_beta,xrp_beta,zec_beta,usdt_beta,dash_beta]
beta_list
```
